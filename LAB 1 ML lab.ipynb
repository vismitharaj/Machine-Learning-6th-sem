```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
# If extracted already
df = pd.read_csv("/content/Dataset of Diabetes .csv")

# View first few rows
df.head()
```

```python
df.isnull().sum()
```

```python
df.drop(['ID','No_Pation'],axis=1,inplace=True)
```

```python
df['Gender']=df['Gender'].map({'M':1,'F':0})
```

```python
df['CLASS'].unique()
```

```python
df['CLASS'] = df['CLASS'].map({'N':0, 'Y':1, 'P':1})
```

```python
def remove_outliers_iqr(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    return df[(df[col] >= lower) & (df[col] <= upper)]


num_cols = ['AGE','Urea','Cr','HbA1c','Chol','TG','HDL','LDL','VLDL','BMI']

# keep original
df_clean = df.copy()

print("Initial shape:", df_clean.shape)

for col in num_cols:
    before = df_clean.shape[0]
    df_clean = remove_outliers_iqr(df_clean, col)
    after = df_clean.shape[0]
    print(f"{col}: removed {before - after} rows")

print("Final shape:", df_clean.shape)
df_clean.head()
df_clean.describe()
```

```python
from sklearn.preprocessing import MinMaxScaler

# numeric columns to scale
num_cols = ['AGE','Urea','Cr','HbA1c','Chol','TG','HDL','LDL','VLDL','BMI']

scaler = MinMaxScaler()

df_clean[num_cols] = scaler.fit_transform(df_clean[num_cols])

df_clean.head()
```

```python
from sklearn.preprocessing import StandardScaler

# numeric columns to scale
num_cols = ['AGE','Urea','Cr','HbA1c','Chol','TG','HDL','LDL','VLDL','BMI']

scaler = StandardScaler()

df_clean[num_cols] = scaler.fit_transform(df_clean[num_cols])

df_clean.head()
```
